### A Simple Guide to using the nQubitTomography Library

Import the main Tomography module

```python
from Tomography import Tomo
import numpy as np
```

Initialize a tomography object with the amount of qubits you'll be performing tomography on.

```python
n_qubits = 2
t = Tomo(n_qubits)
```

You can either import data through an excel file, as shown below. To see how to format your files, please check out example.xlsx.

```python
rho = t.polarization_tomography_MLE(filename="example.xlsx")
```

Or, you can hardcode your data in, as shown below.

```python
projections = ["HH","HV","HD","HR","VH","VV","VD","VR","DH","DV","DD","DR","RH","RV","RD","RR"]
counts = np.array([10,1804,1101,912,1937,13,885,928,908,831,114,
                   1506,970,859,419,216])

rho = t.polarization_tomography_MLE(projections, counts)
```

You can now grab features, such as the density plot, of your tomography object:

```python
t.density_plot(rho.matrix)
```

And format the matrix to better understand the specifics.

```python
t.display_rho(rho.matrix)
```

You can also fetch properties of the resulting matrix from the Rho class:

```python
rho.concurrence
```

```python
rho.purity
```
